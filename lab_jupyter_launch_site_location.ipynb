```python
# TASK 3: Calculate the distances between a launch site to its proximities
import numpy as np
def haversine(lat1, lon1, lat2, lon2):
    # Convert latitude and longitude from degrees to radians
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    
    # Haversine formula
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat / 2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2.0)**2
    c = 2 * np.arcsin(np.sqrt(a))
    r = 6371  # Radius of Earth in kilometers
    return c * r

# Example proximities (you can replace with actual points of interest)
proximities = {
    "Name": ["Nearby City 1", "Nearby Airport 1", "Nearby Port 1"],
    "Latitude": [29.2108, 28.4294, 28.6024],  # Replace with actual latitudes
    "Longitude": [-81.0228, -81.3084, -80.6774]  # Replace with actual longitudes
}

# Create a DataFrame for proximities
proximities_df = pd.DataFrame(proximities)

# Iterate through each launch site and calculate distances to each proximity
distances = []
for idx, launch_site in launch_sites_df.iterrows():
    site_lat, site_lon = launch_site['Lat'], launch_site['Long']
    
    # Calculate the distance to each proximity and store it in a dictionary
    for _, proximity in proximities_df.iterrows():
        distance = haversine(site_lat, site_lon, proximity['Latitude'], proximity['Longitude'])
        distances.append({
            'Launch Site': launch_site['Launch Site'],
            'Proximity': proximity['Name'],
            'Distance (km)': distance
        })

# Create a DataFrame to store distances
distances_df = pd.DataFrame(distances)

# Display the calculated distances
distances_df
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
